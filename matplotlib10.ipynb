# savefig

```python
from matplotlib import pyplot as plt #or (import matplotlib.pyplot as plt )
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("line") #to save graph externally
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("line1",dpi=2000) #dpi(dot per inch) #game etlo zoom kri lo save photo fatse nai e
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("line1",dpi=2000,facecolor='g') #facecolor='g' to give graph green color in boundry
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("line12.pdf",dpi=2000,facecolor='g') #"line12.pdf" to make pdf file
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("line112.jpg",dpi=2000,facecolor='g') #"line12.jpg" to make jpg file #its png bydefault
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("linee",dpi=2000,facecolor='g',transparent=True) #if transparent is true it will make background color to green which is given by user #if false dont make any changes
plt.show()
```

```python
x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')

plt.savefig("line2",dpi=2000,facecolor='g',transparent=False,bbox_inches='tight') #to change box inches
plt.show()
```
